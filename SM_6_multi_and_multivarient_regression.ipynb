```python
ax = sns.displot(df.price,bins=50,kde=True)
plt.axvline(x=np.mean(df.price), color="green", label='mean')
plt.axvline(x=np.median(df.price), color="red", label="meadian")
plt.legend(loc="upper right")
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10))
sns.histplot(data=df,x="car_ID",kde=True,color="red",ax=axs[0,0])
sns.histplot(data=df,x="horsepower",kde=True,color="skyblue",ax=axs[0,1])
sns.histplot(data=df,x="carbody",kde=True,color="olive",ax=axs[1,0])
sns.histplot(data=df,x="enginesize",kde=True,color="gold",ax=axs[1,1])
```

```python
ax=sns.barplot(x="fueltype", y="price", data=df)
```

```python
df['compressionratio'].plot(kind='box',title = 'compressionratio', figsize = (4,4))
```

```python
car_df=df[[ 'car_ID', 'wheelbase', 'carlength', 'carwidth','carheight']]

sns.pairplot(car_df, diag_kind='kde')
plt.show()
```
